```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lets load the dataset
netflix=pd.read_csv('C:/Users/adity/Downloads/netflixdabasic/netflix_titles.csv')
```

```python
netflix
```

```python
#datseinfo
netflix.info()
```

```python
netflix.reset_index(drop=True)
```

```python
#removal of duplicate values
netflix.dropna(inplace=True)
```

```python
netflix
```

```python
#duplicateentries deleted.Lets see new data
netflix.info()
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce', infer_datetime_format=True)
netflix.info()
```

```python
#coloumn types
types=netflix['type'].value_counts().reset_index()
types.columns = ['type', 'count']
types
```

```python
# director coloumn
directors = netflix['director'].value_counts().reset_index()
directors
```

```python
#country
countries = netflix['country'].value_counts().reset_index()
countries
```

```python
#Types of Director
directors = netflix.groupby(['director' , 'type'],as_index=False)['director'].value_counts()
directors
```

```python
#LEts find top 10 directors
top_10_directors = netflix.groupby(['director','type'])['director'].value_counts().sort_values(ascending=False).iloc[0:11]
top_10_directors
```

```python
countries = netflix.groupby(['country','type'],as_index=False)['country'].value_counts()
countries
```

```python
#lets see top 10 countries
top_10_countries = netflix.groupby(['country','type'])['country'].value_counts().sort_values(ascending=False).iloc[0:9]
top_10_countries
```

```python
top_10_Tvshows_countries = countries[countries['type'] == "TV Show"].sort_values(by = 'count',ascending = False).iloc[0:10]
top_10_Tvshows_countries
```

```python
top_10_movie_countries = countries[countries['type'] == "Movie"].sort_values(by='count',ascending = False).iloc[0:10]
top_10_movie_countries
```

```python
#release year
release_years = netflix.groupby(['release_year','type'],as_index=False)['release_year'].value_counts()
release_years
```

```python
top_10_movie_years = release_years[release_years['type'] == 'Movie'].sort_values(by='count',ascending=False)
top_10_movie_years
```

```python
top_10_TV_showyears = release_years[release_years['type'] == "TV Show"].sort_values(by='count',ascending=False)
top_10_TV_showyears 
```

```python
ratings = netflix.groupby(['rating','type'],as_index=False)['type'].value_counts()
ratings
```

```python
top_10_movie_ratings = ratings[ratings['type']=='Movie'].sort_values(by='count',ascending=False).iloc[0:10]
top_10_movie_ratings
```

```python
top_10_webshow_rating = ratings[ratings['type'] == "TV Show"].sort_values(by='count',ascending=False).iloc[0:10]
top_10_webshow_rating 
```

```python
durations = netflix.groupby(['duration','type'],as_index=False)['duration'].value_counts()
durations
```

```python
top_10_movie_durations = durations[durations['type'] == 'Movie'].sort_values(by='count',ascending = False).iloc[0:10]
top_10_movie_durations 
```

```python
top_10_TV_Shows_duration = durations[durations['type'] == 'TV Show'].sort_values(by='count',ascending=False).iloc[0:10]
top_10_TV_Shows_duration 
```

```python
#different genres of movies and webshow

listed_ins = netflix.groupby(['listed_in','type'],as_index=False)['listed_in'].value_counts()
listed_ins
```

```python
#top10 genres for movies
top_10_genres_movies = listed_ins[listed_ins['type'] == 'Movie'].sort_values(by="count",ascending=False).iloc[0:10]
top_10_genres_movies
```

```python
#top 10 genres for TV SHows
top_10_genres_TV_Shows= listed_ins[listed_ins['type'] == 'TV Show'].sort_values(by='count',ascending=False).iloc[0:10]
top_10_genres_TV_Shows
```

```python
#Now it was all about just tables .Lets visualize the data 

types.set_index('type', inplace=True)
types.plot.pie(y = 'count', autopct='%.2f%%', legend='type')
plt.title("Moives VS TV Shows")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
top_10_directors.plot(x='director', y='count', kind='bar')
plt.title('Top 10 Directors')
plt.xlabel('Director')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
#visualizing the country column
plt.figure(figsize=(20, 8))
top_10_countries.plot(x='country', y='count', kind='bar')
plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
netflix['release_year'].plot(kind='hist', bins=30)
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
top_10_webshow_rating.plot(x='rating', y='count', kind='bar')
plt.title('Top 10 TV Shows Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_movie_durations.plot(x='duration', y='count', kind='bar')
plt.title('Top 10 Movie Durations')
plt.xlabel('Duration')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_TV_Shows_duration.plot(x='duration', y='count', kind='bar')
plt.title('Top 10 TV Shows Durations')
plt.xlabel('Duration')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
top_10_genres_movies.plot(x='listed_in', y='count', kind='bar')
plt.title('Top 10 Listed Genres for Movies')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
